```python
import matplotlib.pyplot as plt
```

```python
plt.plot([2,4,9,7])
```

```python
plt.plot([2,4,9,7])
plt.ylabel('Numbers')
plt.xlabel('Indices')
plt.title('Graph Name')
plt.show()
```

```python
# we can also add grid on the graph
plt.plot([2,4,9,7])
plt.ylabel('Numbers')
plt.xlabel('Indices')
plt.grid()
plt.title('Graph Name')
plt.show()
```

```python
plt.plot([1,2,3,4],[1,4,9,16])  # xlabel, ylabel
plt.ylabel('Squares')
plt.xlabel('Numbers')
plt.grid()
```

```python
# putting dots instead of lines
plt.plot([1,2,3,4],[1,4,9,16], 'ro')  # xlabel, ylabel  -- r for red color
plt.ylabel('Squares')
plt.xlabel('Numbers')
plt.grid()
plt.show()
```

```python
import numpy as np
```

```python
t = np.arange(0,5,0.2)
t
```

```python
len(t)
```

```python
# blue dashes, red squares, green triangles
plt.plot(t, t**2, 'b--', label='^2')
plt.plot(t, t**2.2, 'gs', label='^2.2')  # square
plt.plot(t, t**2.5, 'g^', label='^2.5') # triangle

plt.legend()  # legend based on line labels

plt.grid()
plt.show()
```

```python
# making line wider
x = np.arange(10, 20, 2)
y = x**2 
plt.plot(x,y, linewidth=5.0)

plt.grid()
plt.show()
```

```python
x1 = [2,3,4,5,6]
y1 = [2,6,4,10,6]
x2 = [2,3,4,12,6]
y2 = [2,3,7,13,6]

lines = plt.plot(x1, y1, x2, y2)

plt.setp(lines[0], color = 'r', linewidth=2.0, label='x1-y1')
plt.setp(lines[1], color = 'g', linewidth=2.0, label='x2-y2')
plt.legend()
plt.grid()
plt.show()
```
